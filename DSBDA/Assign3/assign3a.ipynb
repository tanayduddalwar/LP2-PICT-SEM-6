```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("nba.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["College"]=dataset["College"].fillna(dataset["College"].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
dataset["Salary"]=dataset["Salary"].fillna(dataset["Salary"].mean())
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.dtypes
```

```python
dataset["Height"].describe()
```

```python
dataset["Height"].value_counts()
```

```python
heightgrp=dataset.groupby(dataset["Height"])
heightgrp.head()
```

```python
for height, group in heightgrp:
    print(f"\nHeight Group: {height}")
    print(group.head()) 
```

```python
heightgrp["Salary"].describe()
```

```python
agegrp=dataset.groupby(dataset["Age"])
agegrp.head()
```

```python
agegrp["Age"].describe()
```

```python
dataset["Age"].min()
```

```python
dataset["Age"].max()
```

```python
bins=[19,25,31,36,40]
ranges=["19-24","25-30","31-35","36-40"]
dataset["AgeGroup"]=pd.cut(dataset["Age"],bins=bins,labels=ranges)
```

```python
dataset["AgeGroup"].head(5)
```

```python
dataset["AgeGroup"].value_counts()
```

```python
age_grp=dataset.groupby(dataset["AgeGroup"])
```

```python
age_grp["Salary"].describe()
```
