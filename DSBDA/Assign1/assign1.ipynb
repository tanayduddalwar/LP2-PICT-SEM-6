```python
data['Age']=2025-data['YearBuilt']
data.drop("YearBuilt",axis=1)
data.to_csv("nona.csv")
```

```python
data.head(5)
```

```python
cat_features=['Suburb','Address','Type','Method','SellerG','Postcode','CouncilArea','Regionname']
data[cat_features]=data[cat_features].astype('category')
```

```python
cat_columns=data.select_dtypes(['category']).columns
data[cat_columns]=data[cat_columns].apply(lambda x:x.cat.codes)
```

```python
data.shape
```

```python
cols_to_keep=['Rooms','Distance','Landsize','Bathroom','Car','BuildingArea','Lattitude','Longtitude','Age','Price']
data_new=data[cols_to_keep]
print(data_new.head(3))
```

```python
data_new.isnull().sum()
```

```python
data_new.describe()
```

```python
y=data_new['Price']
x=data_new.drop(['Price'],axis=1)
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=56)
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_absolute_error as mae
```

```python
lr=LR()
lr.fit(train_x,train_y)
```

```python
train_predict=lr.predict(train_x)
k=mae(train_predict,train_y)
print("Training Mean Absolute Error",k)
```

```python
test_predict=lr.predict(test_x)
k=mae(test_y,test_predict)
print("Test Mean Absolute Error",k)
```

```python
lr.coef_
```

```python
residuals=pd.DataFrame({'fitted values':test_y,'predicted values':test_predict})
residuals['residuals']=residuals['fitted values']-residuals['predicted values']
residuals.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Create a pipeline: first scale, then train
lr = make_pipeline(StandardScaler(), LinearRegression())

# Fit the model
lr.fit(train_x, train_y)

# Predict
train_predict = lr.predict(train_x)

# Evaluate
k = mae(train_predict, train_y)
print('Training Mean Absolute Error:', k)

test_predict = lr.predict(test_x)
k = mae(test_predict, test_y)
print('Test Mean Absolute Error:', k)
```
